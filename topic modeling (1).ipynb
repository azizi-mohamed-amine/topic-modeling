```python
#import packages
import pandas as pd
import pickle
from nltk import word_tokenize, pos_tag
import gensim
from nltk import sent_tokenize

from gensim.utils import lemmatize
from nltk.corpus import stopwords
```

```python
#load the data
data = pd.read_pickle('pre-processed.pkl')
data
```

```python
#get the tokens
stopped_tokens=data['tokens_no_stop']
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(stopped_tokens)
corpus= [dictionary.doc2bow(text) for text in stopped_tokens]
```

```python
import pickle
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

```python
#use LDA
import gensim
NUM_TOPICS = 5  # number of topic to choose
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
ldamodel.save('model5.gensim')
```

```python
topics = ldamodel.print_topics(num_words=4) #number of words represent the topics
for topic in topics:
    print(topic)
```
